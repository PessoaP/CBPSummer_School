```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
potentialE=np.loadtxt("ns/potentialE.dat")
# potentialE=np.loadtxt("/content/drive/MyDrive/summarSchool/potentialE.dat")
```

```python
potentialE.shape
```

```python
plt.figure(figsize=(12,8))
plt.plot(potentialE,color='red', label='E')
#plt.plot(potentialE,'ob')
#########################################################################
plt.xlabel("Simulation steps ",fontsize=25)
plt.ylabel(r"Energy (eV)",fontsize=25)
plt.title("Plot of Energy",fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc=0,fontsize=20)
plt.savefig("./energy.png", dpi=200 )
plt.savefig("./energy.pdf", dpi=200 )
```

```python
plt.figure(figsize=(16,8))
plt.hist(potentialE,bins=100,histtype='stepfilled',density = True,color="g",alpha=0.8,label='$Energy$')
################################################################################################################
plt.xlabel(r"$ X $ (eV)",fontsize=35)
plt.ylabel(r"P($X$)",fontsize=35)
plt.yticks(fontsize=30)
plt.legend(loc=2,frameon=False,fontsize=25)
plt.ylim([0.,1.0])
plt.xlim([-3.2,0.0])
plt.xticks(fontsize=30)
plt.savefig("./Distribution-energy.png", dpi=200 )
plt.savefig("./Distribution-energy.pdf", dpi=200 )
```
