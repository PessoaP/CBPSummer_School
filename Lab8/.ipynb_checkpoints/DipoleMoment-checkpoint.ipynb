# Dipole moments

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Maxwell distribution function 
def pM(x, sigma):
    return (4 * np.pi * x**2 * np.exp(-x**2 / (2 * sigma**2))) / (np.sqrt(2 * np.pi* sigma**2)**3)
```

```python
# Load the dipole moment file, which contains the dipole moment vectors (mu_X, mu_Y, mu_Z) for each frame
d = np.loadtxt("muProtein_gfp_6.dat")
print("shape of dipole moment file:", d.shape)

# Calculate the average dipole moment vector over all frames
# mav represents the mean vector of dipole moments (mu_X, mu_Y, mu_Z) across all frames
mav = np.mean(d, axis=0)

# Calculate the magnitude of the dipole moment for each frame
# dmag is an array where each element represents the magnitude of the dipole moment vector for a specific frame
# dmag= (mu_X^2+mu_Y^2+ mu_Z^2)^0.5 
dmag = np.linalg.norm(d, axis=1)
```

```python
# Get the number of frames (length of the dipole moment array)
# nd represents the total number of frames (or time steps) in the simulation
nd = np.shape(d)[0]
print(nd)

# Initialize an array d2 to store the square of the deviations of dipole moments from the mean dipole moment
# d2[i] will eventually store the squared difference between the dipole moment of the ith frame and the mean dipole moment
d2 = np.zeros(nd)
for i in range(0, nd):
    # Calculate the square of the deviation of the dipole moment vector for each frame from the mean dipole moment
    # This essentially computes (mu_X - <mu_X>)^2 + (mu_Y - <mu_Y>)^2 + (mu_Z - <mu_Z>)^2 for each frame
    d2[i] = np.dot(d[i] - mav, d[i] - mav)

###########
# Calculate the magnitude of the deviation vector (dn) for each frame
# dn is the square root of d2, representing the magnitude of deviation from the mean dipole moment
dn = np.sqrt(d2)

# Calculate the maximum deviation (mmax), set minimum deviation (mmin), define number of bins (n_bins) and bin width (w_bin)
# mmax represents the maximum magnitude of the deviation of the dipole moment
# n_bins is the number of bins to be used in the histogram
# w_bin is the width of each bin in the histogram
mmax = np.max(dn)
mmin = 0
n_bins = 200
w_bin = (mmax - mmin) / n_bins
print(mmax)
print("width of each bin", w_bin)

# Calculate the second moment <M^2> of the deviation distribution
# m2 represents the average of the square of the deviations (mean square deviation)
m2 = np.mean(dn**2)
print(f"<M^2> = {m2} D^2")

############################################################################################
# Calculate the standard deviation (std) of the deviation distribution
# The standard deviation is calculated as the square root of m2 divided by 3
# std represents the spread of the deviation magnitudes
std = np.sqrt(m2 / 3)
print(f"std deviation = {std} D")
```

# Plotting maxwell distribution of dipole moment

```python
# Create a histogram of the deviation magnitudes
# bin_counts represents the number of occurrences in each bin
# bin_edges represents the edges of the bins
bin_counts, bin_edges = np.histogram(dn, bins=n_bins, range=(mmin, mmax))

# Normalize the histogram bin counts to obtain a probability density function (PDF)
# norm is the normalization factor, which ensures that the area under the histogram equals 1
norm = np.sum(bin_counts) * w_bin

# Normalize bin counts to form a probability density function (PDF)
# N_bin_counts is the normalized histogram, representing the probability density of each bin
N_bin_counts = bin_counts / norm

############################################################################################
# Plot the normalized histogram (PDF) as a function of the dipole moment deviation
plt.figure(figsize=(12, 8))
plt.plot(bin_edges[1:], N_bin_counts, label=r'Raw data', color='b', markersize=5, marker='o', linestyle=' ')
plt.xlabel('Dipole Moment (D)', fontsize=25)
plt.ylabel('Probability Density', fontsize=25)
plt.title('Maxwell distribution of Dipole Moments', fontsize=25)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.show()
```

# Fitting of Maxwell function to Maxwell distribution of Dipole Moments

```python
var=np.var(dmag)
guess=np.array([np.sqrt(var)])
popt, pcov = curve_fit(pM,bin_edges[1:],N_bin_counts,guess)
initial_sigma=popt[0]
hist_fit=pM(bin_edges[1:],initial_sigma)
plt.figure(figsize=(10,8))
plt.plot(bin_edges[1:], N_bin_counts,label=r'$\mu$',color='b',markersize=6,marker='o',linestyle=' ')
plt.plot(bin_edges[1:], hist_fit, label=r'Fitted to $\mu$',color='r')
#plt.xlabel(r"$\Delta E$ (eV)",fontsize=20)
#plt.ylabel(r"P($\Delta E$)",fontsize=20)
plt.xlabel('$\mu$ (D)',fontsize=30)
plt.ylabel(r"P($\mu$)",fontsize=30)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.legend(fontsize=20,frameon=False)
plt.show()
```
